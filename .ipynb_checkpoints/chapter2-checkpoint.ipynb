```python
# in this case (2, 3). the \n character creates a new line.
x = np.array([1, 2, 3, 4, 5, 6])
print('beginning x:\n', x)
x_reshape = x.reshape((2, 3))
print('reshaped x:\n', x_reshape)
```

```python
# access the top left element of x_reshape, we type in x_reshape[0,0].
x_reshape[0, 0]
```

```python
# Similarly, x_reshape[1,2] yields the element in the second row and the third column of x_reshape.
x_reshape[1, 2]
```

```python
# modify the top left element of x_reshape
print('x before we modify x_reshape:\n', x)
print('x_reshape before we modify x_reshape:\n', x_reshape)
x_reshape[0, 0] = 5
print('x_reshape after we modify its top left element:\n',
x_reshape)
print('x after we modify top left element of x_reshape:\n', x)
```

```python
# Can we also modify a tuple? It turns out that we cannot
my_tuple = (3, 4, 5)
my_tuple[0] = 2
```

```python
# T provides its transpose.
x_reshape.shape , x_reshape.ndim , x_reshape.T
```

```python
# Generate random data: generate 50 independent random variables from a N(0, 1) distribution.
x = np.random.normal(size=50)
x
```

```python
# create an array y by adding an independent N(50, 1) random variable to each element of x.
y = x + np.random.normal(loc=50, scale=1, size=50)

np.corrcoef(x, y)
```

```python
# each time we call np.random.normal(), we will get a different answer, as shown in the following example
print(np.random.normal(scale=5, size=2))
print(np.random.normal(scale=5, size=2))
```

```python
# To generate normal data we use rng.normal().
rng = np.random.default_rng (1303)
print(rng.normal(scale=5, size=2))
rng2 = np.random.default_rng (1303)
print(rng2.normal(scale=5, size=2))
```

```python
# np.mean(), np.var(), and np.std() functions can be used to compute np.mean() np.var() np.std() the mean
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
np.mean(y), y.mean()
```

```python
np.var(y), y.var(), np.mean((y - y.mean())**2)
```

```python
# Notice that by default np.var() divides by the sample size n rather than n âˆ’ 1; see the ddof argument in np.var?
np.sqrt(np.var(y)), np.std(y)
```

```python
# The np.mean(), np.var(), and np.std() functions can also be applied to the rows and columns of a matrix
X = rng.standard_normal((10, 3))
X
```

```python
# Since arrays are row-major ordered, the first axis, i.e. axis=0, refers to its rows. We pass this argument into the mean() method for the object X.
X.mean(axis=0)
```

```python
# The following yields the same result.
x.mean(0)
```
